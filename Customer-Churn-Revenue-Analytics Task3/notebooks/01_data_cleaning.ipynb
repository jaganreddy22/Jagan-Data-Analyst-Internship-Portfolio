```python
import numpy  as np
import pandas as pd

data=pd.read_csv(r"C:\Users\jagan\Downloads\Customer-Churn-Revenue-Analytics\data\churnrawdata.csv")
df=pd.DataFrame(data)
print(df.head(10))
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
(df['totalcharges'] == " ").sum()
```

```python
df[df['tenure'] == 0]
```

```python
df['totalcharges'].nunique()
```

```python
df['totalcharges'] = df['totalcharges'].replace(" ", 0)
```

```python
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
```

```python
df.info()
```

```python
print(df.duplicated().sum())
```

```python
churn_yes_count=(df['churn']=='Yes').sum()
print(churn_yes_count)
```

```python
churn_rate=(churn_yes_count/df['customerid'].count())*100
print(churn_rate)
```

```python
contract_counts = df.groupby('contract')['customerid'].count()
print(contract_counts)
```

```python
churn_per_contract=df.groupby('churn')['contract'].size().reset_index(name='count')
print(churn_per_contract)
```

```python
churn_table = pd.crosstab(df['contract'], df['churn'])

churn_percentage = churn_table.div(churn_table.sum(axis=1), axis=0) * 100

print(churn_percentage)
```

```python
bins = [0, 12, 24, 48, 72]
labels = ['0-12 Months', '12-24 Months', '24-48 Months', '48+ Months']

df['tenuregroup'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=False)
df['tenuregroup'].value_counts()
```

```python
churn_by_tenure_group=pd.crosstab(df['tenuregroup'],df['churn'])
churn_by_tenure_group_percentage=churn_by_tenure_group.div(churn_by_tenure_group.sum(axis=1),axis=0)*100
print(churn_by_tenure_group_percentage)
```

```python
average_monthly_charges=df.groupby('churn')['monthlycharges'].mean()
average_monthly_charges
```

```python
average_total_charges=df.groupby('churn')['totalcharges'].mean()
average_total_charges
```

```python
payment_churn_table = pd.crosstab(df['paymentmethod'], df['churn'])

payment_churn_percentage = payment_churn_table.div(
    payment_churn_table.sum(axis=1), axis=0
) * 100

print(payment_churn_percentage)
```

```python
import matplotlib.pyplot as plt

contract_churn = pd.crosstab(df['contract'], df['churn'])
contract_churn_percentage = contract_churn.div(contract_churn.sum(axis=1), axis=0) * 100

contract_churn_percentage['Yes'].plot(kind='bar')
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Percentage")
plt.xlabel("Contract Type")
plt.show()
```

```python
churn_by_tenure_group=pd.crosstab(df['tenuregroup'],df['churn'])
churn_by_tenure_group_percentage=churn_by_tenure_group.div(churn_by_tenure_group.sum(axis=1),axis=0)*100
churn_by_tenure_group_percentage['Yes'].plot(kind='bar')
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Percentage")
plt.xlabel("Tenure Group")
plt.show()
```

```python
payment_churn = pd.crosstab(df['paymentmethod'], df['churn'])
payment_churn_percentage = payment_churn.div(payment_churn.sum(axis=1), axis=0) * 100

payment_churn_percentage['Yes'].plot(kind='bar')
plt.title("Churn Rate by Payment Method")
plt.ylabel("Churn Percentage")
plt.xlabel("Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python
average_churn=df.groupby('churn')['monthlycharges'].mean()
average_churn.plot(kind='bar')
plt.title("Average Monthly Charges by Churn")
plt.ylabel("Average Monthly Charges")
plt.show()
```
